## Task-1:Prediction Using Decision Tree Algorithm

```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

### Reading Data

```python
data=pd.read_csv("datasets/Iris (3).csv")
data
```

```python
data.info()
```

```python
x= data[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].values
x[0:5]
```

```python
Y = data['Species']
Y[0:5]
```

```python
from sklearn.model_selection import train_test_split
x_trainset, x_testset, Y_trainset, Y_testset = train_test_splittrain_x, test_x, trian_y, test_y = train_test_split(x,Y,test_size=0.3, random_state=0)
```

```python
SpeciesTree = DecisionTreeClassifier(criterion = 'entropy', max_depth = 4)
SpeciesTree
```

```python
SpeciesTree.fit(x_trainset, Y_trainset)
```

### Prediction

```python
predTree = SpeciesTree.predict(x_testset)
```

```python
print(predTree[0:5])
print(Y_testset[0:5])
```

### Evaluation

```python
from sklearn import metrics
print("DecisionTree's Accuracy:",metrics.accuracy_score(Y_testset, predTree))
```
